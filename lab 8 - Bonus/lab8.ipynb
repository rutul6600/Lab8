```python
#Importiing required libraries to visualize the data from the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mobilephones.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#First Graph
plt.figure(figsize=(12, 4))
sns.countplot(x= df['form'], order=df.form.value_counts().sort_values(ascending = False).index)
plt.title('Types of mobile phone form, sold most')
plt.savefig("Chart 1.png")
```

```python
#Second Graph
plt.figure(figsize=(12, 6))
sns.countplot(x=df['smartphone'])
plt.title('kind of mobile which are sold most')
plt.savefig("Chart 2.png")
```

```python
#Third Graph
plt.figure(figsize=(12, 7))
plt.xticks(rotation = 'vertical')
sns.countplot(x=df['manufacturer'], order=df.manufacturer.value_counts().sort_values (ascending = True).index)
plt.title('Companies that sold the most')
plt.xticks()
plt.savefig("Chart 3.png")
```

```python
apple = df.loc[df.manufacturer == 'Apple']
apple
```

```python
#Fourth Graph
plt.figure(figsize = (10, 5))
plt.xticks(rotation = 'vertical')
sns.barplot(x='model', y ='units_sold_m', data = apple)
plt.title('Apple Phone models which are sold most')
plt.savefig("Chart 4.png")
```

```python
samsung = df.loc[df.manufacturer == 'Samsung']
samsung
```

```python
#Fifth Graph
plt.figure(figsize = (10, 5))
plt.xticks(rotation = 'vertical')
sns.barplot(x='model', y ='units_sold_m', data = samsung )
plt.title('Samsung phone model which are sold most')
plt.savefig("Chart 5.png")
```
